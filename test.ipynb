```python
# target function
# f(x)=x^2 + y^2
```

```python
import random,math

x=random.randint(10,100)
y=random.randint(10,100)
rate=0.1
iteration=1
while x**2+y**2 > 0.01:
    delta_x=2*x
    delta_y=2*y
    nor=math.sqrt(delta_x**2+delta_y**2)
    x-=rate*delta_x/nor
    y-=rate*delta_y/nor
    print(f'loop {iteration}: x={x:.2f} y={y:.2f} f={x**2+y**2:.2f}')
    iteration+=1
```
